```python
#Identificación de Datos Faltantes y Atípicos

plt.figure(figsize=(6, 3))
sns.boxplot(x=Datos['km_driven'])
plt.title('Identificación de Datos Faltantes y Atípicos', fontsize=10 )
```

```python
#Identificación de Datos Faltantes y Atípicos

plt.figure(figsize=(6, 3))
sns.boxplot(x=Datos['Asientos'])
plt.title('Identificación de Datos Faltantes y Atípicos', fontsize=10 )
```

```python

#Revisión de datos faltantes

print("Cantidad de Registros en cero Año : ",Datos['Año'].isin([0]).sum())
print("Cantidad de Registros en cero precio de venta: ",Datos['precio de venta'].isin([0]).sum())
print("Cantidad de Registros en cero km_driven: ",Datos['km_driven'].isin([0]).sum())
print("Cantidad de Registros en cero kilometraje: ",Datos['kilometraje'].isin([0]).sum())
print("Cantidad de Registros en cero máximo pode: ",Datos['máximo poder'].isin([0]).sum())
print("Cantidad de Registros en cero Asientos: ",Datos['Asientos'].isin([0]).sum())
```

```python
#Conteo de Datos Atípicos

nivel_minimo=2
nivel_maximo=10
Total_Atipicos=((Datos['Asientos']<nivel_minimo)|(Datos['Asientos']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'kilometraje', y = 'precio de venta', data = Datos)
plt.subplot(2,3,2)
sns.boxplot(x = 'km_driven', y = 'precio de venta', data = Datos)
plt.subplot(2,3,3)
sns.boxplot(x = 'Año', y = 'Asientos', data = Datos)
plt.subplot(2,3,4)
plt.show()
```

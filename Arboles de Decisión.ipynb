### Datos atipicos

```python
def Datos_atipicos(dataset,num_x,num_y):
    columns_dt =dataset.columns

    # Creacion de la grafica 
    fig, ax = plt.subplots(num_x, num_y, figsize=(30, 20))  
    # ajusta el tamaño de la figura según tus necesidades
    count = 0
    for x in range(0, num_x):
        for y in range(0, num_y):
            try:
                boxes = dataset.groupby(columns_dt[count]).count()[columns_dt[0]]
                sns.boxplot(boxes, ax=ax[x, y])
                ax[x, y].set_title(columns_dt[count])  # agrega un título al subgráfico
                count += 1
            except:
                boxes = dataset.groupby(columns_dt[count]).count()[columns_dt[-1]]
                sns.boxplot(boxes, ax=ax[x, y])
                ax[x, y].set_title(columns_dt[count])  # agrega un título al subgráfico
                count +=1
```

```python
dataset.shape
```

```python
Datos_atipicos(dataset,num_x=2,num_y=6)
```

**Conclusion de la fase**

Encontramos muchos datos con valores atipicos, tendremos que evaluar si sera mejor la eliminacion o el intercambio de los valores, aunque esto puede cambiar la distribucion de los datos

```python
def atipicos_hist_box(dataset,list_atipicos:list):
    fig,ax=plt.subplots(len(list_atipicos),2,figsize=(30,20))

    # Mostrar datos
    for x in range(len(list_atipicos)):
        # Grafica de campana de gauss
        sns.histplot(dataset,x=list_atipicos[x],kde=True,ax=ax[x,0])
        # Grafica del boxplor
        sns.boxplot(dataset,x=list_atipicos[x],ax=ax[x,1])
        plt.subplots_adjust(hspace=0.5)  # Ajusta el espaciado vertical


        # agregar un titulo
        ax[x,0].set_title("analisis columna "+list_atipicos[x],fontsize=20,fontweight='bold')

```

```python
atipicos_hist_box(dataset,['fixed acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','sulphates','alcohol'])
```

Ahora que ya tenemos un mejor acercamiento a los datos atipicos, y ya podemos inferir cuales son esos limites donde habria que cortar; sin embargo, hay que definir que tan conveniente es la eliminacion de los datos atipicos, con respecto a la cantidad total de los datos 

```python
eval_list_r ={'fixed acidity':13,'citric acid':0.7,'residual sugar':5,'chlorides':0.15,'free sulfur dioxide':160,'total sulfur dioxide':160,'density':1.30,'sulphates':1.30,'alcohol':13}
eval_list_l ={'fixed acidity':0,'citric acid':0,'residual sugar':0,'chlorides':0,'free sulfur dioxide':0,'total sulfur dioxide':0,'density':0.992,'sulphates':0,'alcohol':0}

total_result_size_r = sum([len(dataset.drop(dataset[dataset[k] < v].index)) for k,v in list(eval_list_r.items())])
total_result_size_l = sum([len(dataset.drop(dataset[dataset[k] > v].index)) for k,v in list(eval_list_l.items())])

total_result_size = total_result_size_r+total_result_size_l
```

```python
total_result_size/len(dataset)
```

**Conclusion del analisis**

Como conclusion encontramos que si eliminamos todos los datos atipicos habremos perdido un **24%** de los datos totales, o lo que seria aproximadamente 384 datos quedandonos con 1214 o el equivalente porcentual de **76%**.

```python
#Eliminacion de datos
for k in list(eval_list_r.keys()):
    dataset = dataset.drop(dataset[dataset[k]>=eval_list_r[k]].index)
    dataset = dataset.drop(dataset[dataset[k] < eval_list_l[k]].index)
```

```python
atipicos_hist_box(dataset,['fixed acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','sulphates','alcohol'])
```

## Analsis de relacion entre variables

Ahora que los datos ya estan listos para un analisis procederemos a evaluar la relacion existente entre la columna objetivo `quality` y las demas columnas; pero antes debemos eterminar si la columna objetivo escategorica o nominal

```python
# Ejemplo de dataset nominal - nominal
sns.jointplot(dataset,x='density',y='alcohol')
```

```python
# ejemplo de dataset nominal - categorico
sns.jointplot(dataset,x='quality',y='alcohol')
```

Como una conclusion tembprana encontramos que la columna de nuestro interes es de tipo categorico, lo que  hace que funciones como la covarianza y la correlacion, tengan mucho significado, por ende, para analizar la relacion existente entre la columna objetivo y las demas columnas es necesario usar el metodo de    ANOVA.

Pero antes veamos como se dirtirbuyen los resultados en el dataset

```python
grupo_calidad = dataset.groupby('quality')['alcohol'].count().reset_index()

plt.figure(figsize=(8,8))
plt.pie(grupo_calidad['alcohol'],labels=grupo_calidad['quality'],autopct="%1.1f%%")
```

Como vemos en  esta grafica, la mayoria de los vinos estudiados son  o bien de categoria 5,6,7; dado que las demas categorias no tiene mucha relevancia

```python
# Relacion de columnas con la columna objetivo
import scipy.stats as stats

#Agrupar los datos por la categoria Quality
grupo_datos =[ dataset[dataset['quality'] == x] for x in list(dataset['quality'].unique())] 


# Creacion de formula ANOVA
f_stat,p_valor = stats.f_oneway( *grupo_datos)

#Evaluar la relacion dada por el ANOVA
col_relacion =[]
for x in range(len(dataset.columns)):
    if p_valor[x] < 0.05:
        print(f"La columna [{dataset.columns[x]}] tiene relacion")
        col_relacion +=[dataset.columns[x]]
```

Con la prueba de ANOVA encontramos que datos tiene relacion con la columna `quality`, donde como siguiente paso sera preparar las columnas con relacionpara la creacion del modelo de prediccion

## Modelo de preidccion `Desicion Tree`

Ahora que ya sabemos que columnas nos son utiles; procederemos a separar los datos en datos de entrenamiento, y datos de prueba. Los cuales usaremos para entrenar y evaluar el modelo

Como modelo usaremos un desicion tree categorico, dado que lo que queremos predecir son categorias, y como metrica de evaluacion usaremos la metrica `f1` la cual nos ayuda a determinar aquellas predicciones erroneas 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import  train_test_split
from sklearn.metrics import  f1_score,accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

### Separacion de los datos

```python
# Definicion de columnas objetivo y columnas de interes
y_data = np.array(dataset['quality'])
x_data = np.array(dataset[col_relacion])


```

```python
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data)
```

### Creacion del modelo


```python
#Creacion del modelo
model = Pipeline([
    ('normalizador',StandardScaler()),
    ('Tree',DecisionTreeClassifier(random_state=12))
])
model
```

```python
#entrenamiento del modelo
model.fit(x_train,y_train)

#Evaluacion del modelo
y_pred=model.predict(x_test)

# Evaluacion del modelo
resultado_eval =[]
for i in range(1,len(y_test)+1):
    f1=f1_score(y_pred=y_pred[:i],y_true=y_test[:i],average='weighted')
    resultado_eval +=[f1]
```

```python
plt.plot(np.arange(len(y_test)),resultado_eval)
plt.title("Certeza del modelo")
plt.xlabel('Tiempo')
plt.ylabel('Valor por el f1_score')
```

```python
fig,ax = plt.subplots(2,2,figsize=(15,6))

sns.scatterplot(x=x_test.T[0],y=x_test.T[1],hue=y_pred,palette='coolwarm',ax=ax[0,0])
sns.histplot(y_pred,ax=ax[0,1],kde=True)
ax[0,0].set_title('Datos predichos por el modelo',fontsize=20)

sns.scatterplot(x=x_test.T[0],y=x_test.T[1],hue=y_test,palette='coolwarm',ax=ax[1,0])
sns.histplot(y_test,ax=ax[1,1],kde=True)
ax[1,0].set_title('Datos reales',fontsize=20)

```

```python
#Testeo con datos de entrnemiento
fig,ax = plt.subplots(2,2,figsize=(15,6))

y_pred = model.predict(x_train)
sns.scatterplot(x=x_train.T[0],y=x_train.T[1],hue=y_pred,palette='coolwarm',ax=ax[0,0])
sns.histplot(y_pred,ax=ax[0,1],kde=True)
ax[0,0].set_title('Datos predichos por el modelo',fontsize=20)

sns.scatterplot(x=x_train.T[0],y=x_train.T[1],hue=y_train,palette='coolwarm',ax=ax[1,0])
sns.histplot(y_train,ax=ax[1,1],kde=True)
ax[1,0].set_title('Datos reales',fontsize=20)
```

Para este caso tenemos que el modelo, tiene una exactitud perfecta, lo que quiere decir, que el modelo comprende a la perfeccion los datos y logra dar un resultado optimo
